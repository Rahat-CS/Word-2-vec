```python
str = page2.extractText()

def process_genre(str):
    s = ''
    for genre in str.split(" "):
        result = []
        for syn in wordnet.synsets(genre):
            for l in syn.lemmas():
                result.append(l.name())
            sentence = " ".join(result)
            s+= sentence
    return s

str = process_genre(str)
print(str)
```
